```python
import os 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rcParams

data_dir = os.getenv("DATADIR")
plt.style.use(data_dir+'bl.mplstyle')
rcParams["mathtext.fontset"] = 'cm'
rcParams["font.size"]= 12
rcParams["axes.labelsize"]= 13
rcParams["xtick.labelsize"]= 11
rcParams["ytick.labelsize"]= 11
rcParams["legend.fontsize"]= 13

df = pd.read_csv(data_dir+'0_SUMS_Catalogs/FinalCatalogs/SUMS_UV_Candidate_Catalog.csv')
VBE = df[df.Ranking == 'VB-E']
VBG = df[df.Ranking == 'VB-G']
BE = df[df.Ranking == 'B-E']
BG = df[df.Ranking == 'B-G']


c1 = '#fe439b' 
c4 = '#72788D'
c3 = '#7A89C2'
c2 = '#E3D7FF'

ranks = ['VB-E', 'VB-G', 'B-E', 'B-G']
rank_colors = [c1,c2,c3,c4]
zorder = [4,3,2,1]

f,ax = plt.subplots(1,1,figsize=(3.35,3.35))

for rank, color,z in zip(ranks, rank_colors, zorder):
    rows = df[df.Ranking == rank]
    #ax.hist(rows['V'], bins=6*4, color=color, alpha=0.9, label=rank, linewidth=0.5,zorder=z, histtype='stepfilled', edgecolor='black', density=True);
    ax.hist(rows['V'], bins=5*3, color=color, alpha=0.8, label=rank, linewidth=0.5,zorder=z, histtype='stepfilled', edgecolor='k');

ax.set_ylabel('N Candidate Stripped Stars',labelpad=1.5)
ax.set_xlabel('V-band Magnitude [AB]',labelpad=1.5)

ax.set_xlim(15.7, 21.5)
xticks = np.arange(16, 21+1, 1)
ax.set_xticks(xticks)
ax.set_xticklabels(xticks, rotation=0)

ax.set_ylim(0, 78)

plt.tight_layout()
ax.legend(frameon=False, loc='upper left', fontsize=10)
plt.savefig('CandidateVmags.pdf', dpi=300)
```
